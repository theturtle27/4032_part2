```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import sklearn
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import scale

import sklearn.metrics as sm
from sklearn import datasets
from sklearn.metrics import confusion_matrix,classification_report
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = 7,4
```

```python
iris = datasets.load_iris()

X = scale(iris.data)

y= pd.DataFrame(iris.target)
variable_names = iris.feature_names
X[0:10,]
```

```python
clustering = KMeans(n_clusters=3, random_state=5)
clustering.fit(X)
clustering_KMeans_plus = KMeans(n_clusters=3, random_state=5, init='k-means++')
clustering_KMeans_plus.fit(X)
```

```python
iris_df = pd.DataFrame(iris.data)
print(iris_df.size)
iris_df.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']
y.columns =['Targets']
```

```python
color_theme = np.array(['darkgray','lightsalmon','powderblue'])
#re_label = np.choose(clustering.labels_,[2,0,1]).astype(np.int64)

plt.subplot(1,2,1)
plt.scatter(x=iris_df.Petal_Length, y=iris_df.Petal_Width, c=clustering.labels_.astype(float), s=20)
plt.title('Ground Truth Classification')
plt.subplot(1,2,2)
plt.scatter(x=iris_df.Petal_Length, y=iris_df.Petal_Width, c=clustering.labels_.astype(int), s=20)
plt.title('K-means Classification')
```

```python
re_label = np.choose(clustering.labels_,[2,0,1]).astype(np.int64)

plt.subplot(1,2,1)
plt.scatter(x=iris_df.Petal_Length, y=iris_df.Petal_Width, c=color_theme[iris.target], s=30)
plt.title('Ground Truth Classification')
plt.subplot(1,2,2)
plt.scatter(x=iris_df.Petal_Length, y=iris_df.Petal_Width, c=color_theme[clustering.labels_], s=30)
plt.title('K-means Classification')
```

```python
plt.subplot(1,2,1)
plt.scatter(x=iris_df.Petal_Length, y=iris_df.Petal_Width, c=clustering.labels_.astype(int), s=20)
plt.title('K-means Classification')
plt.subplot(1,2,2)
plt.scatter(x=iris_df.Petal_Length, y=iris_df.Petal_Width, c=clustering_KMeans_plus.labels_.astype(int), s=20)
plt.title('K-means++ Classification')
```

```python
print(classification_report(y, re_label))
```
